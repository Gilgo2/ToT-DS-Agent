```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
error_dict = {
    'KeyError': 76,
    'ValueError': 21,
    'TypeError': 3,
    'ImportError': 1,
    'AttributeError': 4,
    'SyntaxError': 3,
    'IndexingError': 1,
    'ModuleNotFoundError': 1,
    'ColumnsIntegerError': 25    
}
```

```python
pdf = pd.DataFrame.from_dict(error_dict, orient='index').reset_index()
pdf = pdf.rename(columns={"index":"error", 0:"error_count"})
pdf = pdf.sort_values(by='error_count')
```

```python
plt.barh(pdf['error'], pdf['error_count'])
```
